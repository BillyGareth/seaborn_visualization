# <font color='silver'>Statistical Relationship-relpot()<font>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
b_g=pd.read_csv('train1.csv')
```

```python
b_g
```

```python
sns.relplot(x='Item_Weight',y='Item_Fat_Content',data=b_g)
```

```python
fats_dict={'Low Fat':0,'Regular':1,'low fat':0,'LF':0,'reg':1}
```

```python
b_g['Item_Fat_content']=b_g['Item_Fat_Content'].replace(fats_dict,regex=True)
```

```python
sns.relplot(x='Item_Weight',y='Item_Type',data=b_g)
```

# <font color='sky blue'> Hue Semantic adding dimension

```python
sns.relplot(x='Item_Weight',y='Item_Type',hue='Outlet_Establishment_Year',data=b_g)
```

# <font color = 'sky blue'> line plot

```python
sns.relplot(x='Item_Weight',y='Item_MRP',kind='line',data=b_g)
```
