```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
movie_df=pd.read_csv('MoviesData.csv')
```

```python
movie_df.dtypes
```

```python
movie_df.head(3)
```

```python
movie_df=movie_df.head(10)
```

```python
sns.swarmplot(x='Titles',y='IMDb',data=movie_df)
plt.xticks(rotation=70)
plt.rcParams['xtick.labelsize']=15
```
