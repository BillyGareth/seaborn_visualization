```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
bl=pd.read_excel('AdventureWorks_Database.xlsx')
```

```python
x=sns.PairGrid(bl)
y=x.map(plt.scatter)
```

```python
y=sns.PairGrid(bl)
y=y.map_diag(plt.hist)
y=y.map_offdiag(plt.scatter)
```

```python
y=sns.PairGrid(bl,hue='FiscalYear')
y=y.map_diag(plt.hist)
y=y.map_offdiag(plt.scatter)
```

```python
y=sns.PairGrid(bl,hue='FiscalYear')
y=y.map_diag(plt.hist)
y=y.map_offdiag(plt.scatter)
y=y.add_legend()
```

```python
y=sns.PairGrid(bl,hue='FiscalYear',palette='coolwarm')# palette=['husl','winter_r','RdBu']
y=y.map_diag(plt.hist)
y=y.map_offdiag(plt.scatter)
y=y.add_legend()
```

```python
y=sns.PairGrid(bl,hue='FiscalYear',palette='autumn')# palette=['husl','winter_r','RdBu']
y=y.map_diag(plt.hist,histtype='step',linewidth=4)
y=y.map_offdiag(plt.scatter)
y=y.add_legend()
```
